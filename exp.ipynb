# Power series of exponential function

```python
f = e^x

a = animate([plot(piecewise([((-10,10), f.series(x,i))]),xmin=-4, xmax=4,ymin=-2, ymax=8, legend_label ='n='+str(i),rgbcolor='blue',thickness=2)+plot(e^x,rgbcolor='red',xmin=-4, xmax=4, ymin=-2, ymax=8,thickness=2) for i in range(1,15)])

a.show(delay=80)
```

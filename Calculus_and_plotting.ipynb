# Calculus and plotting with Sage

The variable ``x`` is predefined in Sage to use. You can simply define a 

```python
f(x) = atan(x) # arctan function
```

and then compute derivatives, indefinite and definite integrals:

```python
f.diff(x)
```

```python
f.integral(x)
```

```python
f.integral(x, 0, pi/2)
```

Note that the above calculation is exact (not floating point)

You can also plot the function:

```python
f.plot(xmin=-5, xmax=5)
```

The optional arguments ``xmin`` and ``xmax`` are the X-axis range of the plot.
There are many options for plotting. Please refer to the documentation at http://doc.sagemath.org/html/en/reference/plotting/sage/plot/plot.html.

For example, you may choose a logarithmic scale along one of the axes, and add a legend.

```python
f(x) = 3^x
f.plot(scale="semilogy", base=3, legend_label="Exponential function is linear on a log scale")
```

A Sage plot is a ``Graphics`` object. Such object may consist of several graphics primitives. Several such primitives may be put together to form a single plot:

```python
sum([(x^i).plot(xmin=0, xmax=1, rgbcolor = (1-i/10.0, i/10.0, 0)) for i in range(10)], Graphics())
```

Besides plotting, differentiating, and integrating functions, Sage can compute Fourer series and Fourier coefficients for functions. This is explained in http://doc.sagemath.org/html/en/reference/functions/sage/functions/piecewise.html

In the above codeis functionally equivalent to the following:

```python
G = Graphics() #Create an empty Graphics object
for i in range(10):
    G += (x^i).plot(xmin=0, xmax=1, rgbcolor = (1-i/10.0, i/10.0, 0)) # Add a graphics primitive at each step to G
G # This shows G
```

Further data can be passed at the time of displaying a graphics object, for example, you can turn off the display of axes:

```python
G.show(axes=False)
```

Animated gif's can be created. In the following example we also compute the Fourier expansion of a step function; see http://doc.sagemath.org/html/en/reference/functions/sage/functions/piecewise.html

```python
f(x) = piecewise([((0, pi/2), -1), ((pi/2, pi), +1)])
a = animate([plot(f.fourier_series_partial_sum(i, pi/2), xmin=0, xmax=pi, ymin=-1.3, ymax=1.3, rgbcolor=(1-i/10.0, i/10.0, 0)) for i in range(10)])
```

```python
a.show(delay=50)
```

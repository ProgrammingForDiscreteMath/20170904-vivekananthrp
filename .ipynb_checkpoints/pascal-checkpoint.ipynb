```python
import matplotlib.pyplot as plt


def Pascal(n):
    """
    Returns the matplotlib figure for ``Pascal's triangle`` upto ``n`` rows
    """
    fig, ax = plt.subplots(figsize=(n,n/3.0))
    #fig, ax = plt.subplots()
    
    #Initiate a coordinate dictionary to store position
    coor ={}

    for i in range(n):
        for j in range(i+1):
            val = factorial(i)/(factorial(j)*factorial(i-j))
            if val%2 == 0:
                color = 'b'
            else:
                color ='r'

            if (j,i) == (0,0):
                try:
                    coor[i].append((j,i))
                except KeyError:
                    coor[i] = [(j,i)]
                ax.annotate(str(val),(j,i),(j,i),color=color)
            elif j == 0:
                try:
                    coor[i].append((-i,-i))
                except KeyError:
                    coor[i] = [(-i,-i)]
                ax.annotate(str(val),(-i,-i),(-i,-i),color=color)            
            elif j == i:
                try:
                    coor[i].append((i,-i))
                except KeyError:
                    coor[i] = [(i,-i)]
                ax.annotate(str(val),(i,-i),(i,-i),color=color)
            elif j>0 or j<i:
                x = (coor[i-1][j-1][0] + coor[i-1][j][0])/2
                coor[i].append((x,-i))
                ax.annotate(str(val),(x,-i),(x,-i),color=color)
    
    ax.set_xlim(-(n+2), (n+2))
    ax.set_ylim(-(n+2),0)
    ax.axis('off')
    fig.tight_layout()
    #fig.suptitle('Pascal\'s triangle')
    return fig
```

```python
Pascal(32)
```

```python
f = Pascal(32)
f.savefig('Pascal.png')
```
